```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import Slider
from matplotlib.animation import FuncAnimation
```

```python
X, Y = np.meshgrid(np.arange(-2, 2, 0.2), np.arange(-2, 2, 0.2))

# Initialize U, V with nonzero values
initial_frequency = 1.0
U = np.sin(X * initial_frequency)
V = np.cos(Y * initial_frequency)

fig, ax = plt.subplots()
plt.subplots_adjust(left=0.25, bottom=0.25)
quiver = ax.quiver(X, Y, U, V)

axfreq = plt.axes([0.25, 0.1, 0.65, 0.03])
freq_slider = Slider(axfreq, 'Freq', 0.1, 2.0, valinit=1.0)

def update(val):
    frequency = freq_slider.val
    U = np.sin(X * frequency)
    V = np.cos(Y * frequency)
    quiver.set_UVC(U, V)

freq_slider.on_changed(update)
```
